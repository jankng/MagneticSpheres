```python
#from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
import numpy as np
```

```python
fileName = "dump.txt"
f = open(fileName, "r")
x, y, z, u, v, w = ([] for i in range(0, 6, 1))

for line in f:
    buffer = line.split(" ")
    x.append(float(buffer[1]))
    y.append(float(buffer[2]))
    z.append(float(buffer[3]))
    u.append(float(buffer[4]))
    v.append(float(buffer[5]))
    w.append(float(buffer[6]))
    
print(w)
```

```python
#%matplotlib notebook

fig, ax = plt.subplots()

ax.scatter(x, z, s=600, alpha = 0.5)
ax.quiver(x, z, u, w, color="red")
ax.set(xlim=(-2, 10), ylim=(2, 12))


plt.show()
```

```python
%matplotlib notebook

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlim3d(0, 10)
ax.set_ylim3d(0, 10)
ax.set_zlim3d(1, 5)
ax.quiver(x, y, z, u, v, w, length = 0.5, normalize = True, color="red")


plt.show()
plt.savefig('single_zoom.eps', format='eps', dpi=1000)
```
