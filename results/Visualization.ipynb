```python
#from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
import numpy as np
```

```python
fileName = "dump.txt"
f = open(fileName, "r")
x, y, z, u, v, w = ([] for i in range(0, 6, 1))

for line in f:
    buffer = line.split(" ")
    x.append(float(buffer[1]))
    y.append(float(buffer[2]))
    z.append(float(buffer[3]))
    u.append(float(buffer[4]))
    v.append(float(buffer[5]))
    w.append(float(buffer[6]))
    print(line)
```

```python
#%matplotlib notebook

fig, ax = plt.subplots()

ax.scatter(x, z, s=700, alpha = 0.5, c="grey")
ax.quiver(x, z, u, w, color="red")
ax.set(xlim=(-4, 6), ylim=(92, 102))
plt.gca().set_aspect('equal', adjustable='box')


plt.show()
```
